```python
import requests
import bs4
from bs4 import BeautifulSoup

def all_cases(url):
    r = requests.get(url)
    mypage = BeautifulSoup(r.text)
    mybody = mypage.find('div',attrs={'class':'body'})
    mytable = mybody.find('table',attrs={'class':"list_table"})
    mytr = mytable.find_all('tr')
    mytr = mytr[2:]  # in the list 'tr' , the data after index 2 is what we want, so we made a slice
    data = [] 
    for i in mytr[:-1]:  # here is the biggest issue we met, for all 'td' in 'tr', 
    #they are in the same structure, except for the last one, because the last row of 'td' is the button 'next page', which is not an organized table as above,
    #so we skipped the last index of the 'mytr' list
        year = i.find_all('td')[0].text
        date = i.find_all('td')[1].text
        district = i.find_all('td')[2].text
        address = i.find_all('td')[3].text
        building =i.find_all('td')[4].text.strip()
        case = i.find_all('td')[5].text   
        data.append([year,date,district,address,building,case]) # collect the every single case in a list
    return data

my_data = [] 
for i in range(1, 10):  # collected the data from page 1 to 9 
    url = 'https://www.squarefoot.com.hk/haunted/?page={}'.format(i)  # made a format of url
    mydata = all_cases(url) # called the function
    my_data.extend(mydata)
print(my_data)    
```

```python
import csv
with open('Haunted House in Hong Kong 2014-2018.csv','w') as f:
    writer = csv.writer(f)
    header = ['Year','Date', 'District', 'Address', 'Building', 'Case']
    writer.writerow(header)
    writer.writerows(my_data)     
```
