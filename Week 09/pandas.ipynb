```python
mapping = {
    '$101-200': 200,
    '$201-400': 400,
    '$51-100': 100,
    '$401-800': 800,
    '$50以下': 50,
}
```

```python
mapping
```

```python
def convert(x):
    return mapping[x]
```

```python
convert('$401-800')
```

```python
convert('$50以下')
```

```python
df['price'] = df['price-range'].apply(convert)
```

```python
df.sort_values(by='price')
```
