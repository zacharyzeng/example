## Error 1 : Header error
Solution: rename function

```python
import pandas
df = pandas.read_csv('error_sample.csv')
```

```python
df.head()
```

```python
df.rename(columns = {'price':'location','likes':'price','location':'likes'},inplace=True)
df.head()
```

## Error 2 : Data format error
solution: replace function

```python
df['price'].value_counts()
```

```python
df['price'].replace({'$101.00':'$101-200'}).value_counts()
```

```python
df
```

## Error 3 : Format/Unity error
Solution: mapping & apply

```python
df['country'].value_counts()
```

```python
mapping={
    '意大利':'意大利菜',
    '韩国菜':'韓國菜'
}

def cleaning(c):
    return mapping.get(c,c)
df['country'].apply(cleaning).value_counts()
```
