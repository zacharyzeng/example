```python
data = []
for i in mytr[:-1]:
    year = i.find_all('td')[0].text
    date = i.find_all('td')[1].text
    district = i.find_all('td')[2].text
    address = i.find_all('td')[3].text
    building =i.find_all('td')[4].text.strip()
    case = i.find_all('td')[5].text   
    data.append([year,date,district,address,building,case])
```
